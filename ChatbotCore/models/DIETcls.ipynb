```python
import requests
import json
```

```python
def WriteIntentToTxtFile(sentence, nameFileTxt):
    URL = "http://127.0.0.1:5005/model/parse"
    message = {"text": sentence}
    r = requests.post(url=URL, data=json.dumps(message))
    # textLines = []
    # for intent in r.json()["intent_ranking"]:
    #     print(intent)
    #     nameIntent = intent["name"]
    #     IntentConfidence = intent["confidence"]
    #     textLines.append(f"- {nameIntent} - {IntentConfidence} \n")
    text = r.json()["intent_ranking"][0]
    with open(f"./intentDIETcls/{nameFileTxt}.txt", "a") as f:
        f.writelines(f"Sentence : {sentence} \n")
        f.writelines(f'{str(text)} \n')
        f.close()
```
